```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# --- 1. Load and Preprocess Data ---
# Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Reshape data to include channel dimension (for Conv2D layers)
# Images are 28x28 grayscale, so add a channel dimension of 1
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# One-hot encode the labels
y_train_one_hot = keras.utils.to_categorical(y_train, 10)
y_test_one_hot = keras.utils.to_categorical(y_test, 10)

# Class names for Fashion MNIST
CLASS_NAMES = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# --- 2. Data Augmentation ---
# This will generate augmented images during training
train_datagen = ImageDataGenerator(
    rotation_range=10,         # Random rotations up to 10 degrees
    width_shift_range=0.1,     # Random horizontal shifts
    height_shift_range=0.1,    # Random vertical shifts
    zoom_range=0.1,            # Random zoom
    shear_range=0.1,           # Shear transformations
    horizontal_flip=True,      # Randomly flip images horizontally
    fill_mode='nearest'        # Strategy for filling in new pixels
)

# It's good practice to split a small portion of x_train for validation
# while using the full x_train for augmentation.
# Alternatively, you can explicitly create a validation set before augmentation.
# For simplicity, we'll let Keras' fit method handle the split for validation during training.

# --- 3. Build the Deeper Model with Batch Normalization & GlobalAveragePooling ---
model = models.Sequential([
    # Block 1
    layers.Conv2D(32, (3, 3), padding='same', input_shape=(28, 28, 1)),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.Conv2D(32, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Block 2
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Block 3 (Added for more depth)
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.25),

    # Classifier Head
    layers.GlobalAveragePooling2D(), # Replaces Flatten, acts as strong regularization
    layers.Dense(256),
    layers.BatchNormalization(), # Batch norm on dense layers too
    layers.ReLU(),
    layers.Dropout(0.5), # Increased dropout for dense layer
    layers.Dense(10, activation='softmax')
])

model.summary()

# --- 4. Callbacks for Training ---
# Early Stopping: Stop if validation loss doesn't improve for 10 epochs
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True,
    verbose=1
)

# Reduce Learning Rate on Plateau: Reduce LR if validation loss doesn't improve
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.2, # Reduce learning rate by a factor of 0.2 (i.e., new_lr = old_lr * 0.2)
    patience=4, # Wait for 4 epochs before reducing
    min_lr=0.00001, # Minimum learning rate
    verbose=1
)

# --- 5. Compile the Model ---
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001), # Start with a standard LR
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# --- 6. Train the Model with Augmentation and Callbacks ---
batch_size = 64
epochs = 100 # We will rely on early stopping to prevent overfitting

print("\nStarting model training...")
history = model.fit(
    train_datagen.flow(x_train, y_train_one_hot, batch_size=batch_size),
    epochs=epochs,
    validation_data=(x_test, y_test_one_hot), # Use x_test as validation for EarlyStopping
                                             # For better practice, split x_train into train/val
                                             # but for simplicity here, we use x_test as val_data.
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)
print("Model training finished.")

# --- 7. Evaluate the Model ---
print("\n--- Model Evaluation ---")
test_loss, test_acc = model.evaluate(x_test, y_test_one_hot, verbose=0)
print(f"Final Test Accuracy: {test_acc*100:.2f}%")
print(f"Final Test Loss: {test_loss:.4f}")

# Predict the probabilities for the test set
y_pred_probs = model.predict(x_test)
# Convert probabilities to class labels
y_pred = np.argmax(y_pred_probs, axis=1)

print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred, target_names=CLASS_NAMES))

# --- 8. Visualize Training History ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# --- 9. Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# --- 10. Visualize Misclassified Images (Optional) ---
print("\n--- Examples of Misclassified Images ---")
misclassified_indices = np.where(y_pred != y_test)[0]
num_images = min(15, len(misclassified_indices)) # Show up to 15 misclassified images

if num_images > 0:
    plt.figure(figsize=(15, 8))
    for i, idx in enumerate(np.random.choice(misclassified_indices, num_images, replace=False)):
        plt.subplot(3, 5, i + 1)
        plt.imshow(x_test[idx].reshape(28, 28), cmap='gray')
        plt.title(f"True: {CLASS_NAMES[y_test[idx]]}\nPred: {CLASS_NAMES[y_pred[idx]]}", fontsize=8)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
else:
    print("No misclassified images found (or model is perfect!).")
```
